```python
import matplotlib.pyplot as plt

T1 = 341.9183  # tiempo secuencial
p = [2, 4, 6, 8] # número de procesos/hilos

T_multiproc = [157.4262, 111.0910, 62.9560, 46.5342]
T_hilos     = [390.9230, 350.8735, 335.4182, 336.0818]
T_mpi       = [129.2700,  77.5968,  60.5794,  53.5329]

# Speedup = T1 / Tp
S_multiproc = [T1/t for t in T_multiproc]
S_hilos     = [T1/t for t in T_hilos]
S_mpi       = [T1/t for t in T_mpi]



plt.figure(figsize=(8,6))
plt.plot(p, S_multiproc, marker='o', label='Multiprocessing')
plt.plot(p, S_hilos,     marker='s', label='Hilos (Threads)')
plt.plot(p, S_mpi,       marker='^', label='MPI')

plt.title("Speedup comparativo (T1 / Tp)")
plt.xlabel("Número de procesos / hilos")
plt.ylabel("Speedup")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
